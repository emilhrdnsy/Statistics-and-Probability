### Mengimport Library

```python
%matplotlib inline

import pandas as pd #untuk proses analisis data, mengelola data berbentuk table atau dataframe dan manipulasi data
import matplotlib.pyplot as plt #lib untuk visualisasi data seperti grafik histogram, diagram batang dan jenis diagram lainnya
import datetime #untuk melakukan sesuatu yang menggunakan waktu sepert untuk memanipulasi tanggal, waktu, dan interval waktu. 
```

### Mengimport file dataset

```python
data = pd.read_csv('D:/Unhas/Big Data Analysis/Case Study 1/trip.csv') #Membaca file dataset
```

### Menjabarkan tanggal ke dalam variabel tersendiri

```python
#Mengkonversi String ke datetime, dan memperoleh fitur baru
list_of_starttime = list(data['starttime'])

list_of_starttime = [datetime.datetime.strptime(x, '%m/%d/%Y %H:%M') for x in list_of_starttime]
data['starttime_mod'] = pd.Series(list_of_starttime,index=data.index)
data['starttime_date'] = pd.Series([x.date() for x in list_of_starttime],index=data.index)
data['starttime_year'] = pd.Series([x.year for x in list_of_starttime],index=data.index)
data['starttime_month'] = pd.Series([x.month for x in list_of_starttime],index=data.index)
data['starttime_day'] = pd.Series([x.day for x in list_of_starttime],index=data.index)
data['starttime_hour'] = pd.Series([x.hour for x in list_of_starttime],index=data.index)
```

### a. Tentukan distribusi jumlah perjalanan per tahun. Apakah Anda melihat pola tertentu?

```python
# Distribusi perjalanan per tahun
groupby_year = data.groupby('starttime_year').size()
groupby_year.plot.bar(title = 'Distribution by Year')
print(groupby_year )
```

### b. Tentukan distribusi jumlah perjalanan per bulan. Apakah Anda melihat pola tertentu?

```python
distribution_by_month = data.groupby('starttime_month').size()
distribution_by_month.plot.bar(title = 'Distribution by Month')
print(distribution_by_month )
```

### c. Determine the distribution of number of trips by day. Do you see a specific pattern?

```python
distribution_by_day = data.groupby('starttime_month').size()
distribution_by_day.plot.bar(title = 'Distribution by Day')
print(distribution_by_day)
```

### d. Buat plot distribusi frekuensi perjalanan setiap hari

```python
groupby_day = data.groupby('starttime_date').size()
groupby_day.plot.bar(title = 'frequency distribution of trips on a daily basis', figsize = (15,4))
```
